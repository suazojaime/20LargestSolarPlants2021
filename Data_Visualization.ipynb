```python
import pandas as pd
import folium
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_excel('20 Largest Solar Power Plants.xlsx')
df
```

```python
df_by_country = df.groupby(by= 'Country')['MWp (Megawatts Peak Capacity)'].sum()
df_by_country = pd.DataFrame({'Country':df_by_country.index, 'MWp' : df_by_country.values} )
```

```python

#fig = px.scatter_geo(df, locations="iso_alpha", color="continent",
#                  hover_name="country", size="pop",
#                  projection="natural earth")

fig = px.scatter_geo(df_by_country, color='Country', locations="Country", locationmode='country names',
                     hover_name='Country', size="MWp", text='MWp',
                     projection="natural earth")

fig.update_layout(title='MegaWattspeak by country of top 20 Largest Solar Plants 2021')
fig.show()



```

```python
sns.relplot(x='Sq Km', y='Country', size='MWp (Megawatts Peak Capacity)',
            sizes=(100, 400), alpha=0.8, palette="muted",
            height=10, data=df).set(title='MegaWattspeak by country by SqKm by Solar Plant');
```
